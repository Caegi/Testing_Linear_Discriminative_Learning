```python
import pandas as pd
from datasets import Dataset 
from inflecteur import inflecteur

df = pd.read_csv('/home/caegi/Documents/M1/Morpho/assets/distrib/vlexique.csv').iloc[:10]
dataset = Dataset.from_pandas(df)
pd.set_option('display.max_columns', None)

inflecteur = inflecteur()
inflecteur.load_dict()

def get_ortho(batch):
    l_lexemes = batch["lexeme"]
    
    l_prs_1sg = []
    l_prs_2sg = []
    l_prs_3sg = []
    l_prs_1pl = []
    l_prs_2pl = []
    l_prs_3pl = []
    
    l_ipfv_1sg = []
    l_ipfv_2sg = []
    l_ipfv_3sg = []
    l_ipfv_1pl = []
    l_ipfv_2pl = []
    l_ipfv_3pl = []

    l_fut_1sg = []
    l_fut_2sg = []
    l_fut_3sg = []
    l_fut_1pl = []
    l_fut_2pl = []
    l_fut_3pl = []

    l_cond_1sg = []
    l_cond_2sg = []
    l_cond_3sg = []
    l_cond_1pl = []
    l_cond_2pl = []
    l_cond_3pl = []

    l_sbjv_1sg = []
    l_sbjv_2sg = []
    l_sbjv_3sg = []
    l_sbjv_1pl = []
    l_sbjv_2pl = []
    l_sbjv_3pl = []

    l_pst_1sg = []
    l_pst_2sg = []
    l_pst_3sg = []
    l_pst_1pl = []
    l_pst_2pl = []
    l_pst_3pl = []

    l_pst_sbjv_1sg = []
    l_pst_sbjv_2sg = []
    l_pst_sbjv_3sg = []
    l_pst_sbjv_1pl = []
    l_pst_sbjv_2pl = []
    l_pst_sbjv_3pl = []

    l_imp_2sg = []
    l_imp_1pl = []
    l_imp_2pl = []
    
    for lexeme in l_lexemes:
        if (lexeme is not None):
            l_prs_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Présent', pos="Verbe"))
            l_prs_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Présent', pos="Verbe"))
            l_prs_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Présent', pos="Verbe"))
            l_prs_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Présent', pos="Verbe"))
            l_prs_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Présent', pos="Verbe"))
            l_prs_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Présent', pos="Verbe"))
            
            l_ipfv_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Imparfait', pos="Verbe"))
            l_ipfv_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Imparfait', pos="Verbe"))
            l_ipfv_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Imparfait', pos="Verbe"))
            l_ipfv_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Imparfait', pos="Verbe"))
            l_ipfv_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Imparfait', pos="Verbe"))
            l_ipfv_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Imparfait', pos="Verbe"))
        
            l_fut_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Futur', pos="Verbe"))
            l_fut_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Futur', pos="Verbe"))
            l_fut_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Futur', pos="Verbe"))
            l_fut_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Futur', pos="Verbe"))
            l_fut_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Futur', pos="Verbe"))
            l_fut_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Futur', pos="Verbe"))
        
            l_cond_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Conditionnel', pos="Verbe"))
            l_cond_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Conditionnel', pos="Verbe"))
            l_cond_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Conditionnel', pos="Verbe"))
            l_cond_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Conditionnel', pos="Verbe"))
            l_cond_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Conditionnel', pos="Verbe"))
            l_cond_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Conditionnel', pos="Verbe"))
        
            l_sbjv_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Subjonctif Présent', pos="Verbe"))
            l_sbjv_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Subjonctif Présent', pos="Verbe"))
            l_sbjv_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Subjonctif Présent', pos="Verbe"))
            l_sbjv_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Subjonctif Présent', pos="Verbe"))
            l_sbjv_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Subjonctif Présent', pos="Verbe"))
            l_sbjv_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Subjonctif Présent', pos="Verbe"))
        
            l_pst_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Passé simple', pos="Verbe"))
            l_pst_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Passé simple', pos="Verbe"))
            l_pst_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Passé simple', pos="Verbe"))
            l_pst_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Passé simple', pos="Verbe"))
            l_pst_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Passé simple', pos="Verbe"))
            l_pst_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Passé simple', pos="Verbe"))
        
            l_pst_sbjv_1sg.append(inflecteur.inflect_word(lexeme, number="s", person="1", tense='Subjonctif Imparfait', pos="Verbe"))
            l_pst_sbjv_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Subjonctif Imparfait', pos="Verbe"))
            l_pst_sbjv_3sg.append(inflecteur.inflect_word(lexeme, number="s", person="3", tense='Subjonctif Imparfait', pos="Verbe"))
            l_pst_sbjv_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Subjonctif Imparfait', pos="Verbe"))
            l_pst_sbjv_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Subjonctif Imparfait', pos="Verbe"))
            l_pst_sbjv_3pl.append(inflecteur.inflect_word(lexeme, number="p", person="3", tense='Subjonctif Imparfait', pos="Verbe"))
        
            l_imp_2sg.append(inflecteur.inflect_word(lexeme, number="s", person="2", tense='Impératif Présent', pos="Verbe"))
            l_imp_1pl.append(inflecteur.inflect_word(lexeme, number="p", person="1", tense='Impératif Présent', pos="Verbe"))
            l_imp_2pl.append(inflecteur.inflect_word(lexeme, number="p", person="2", tense='Impératif Présent', pos="Verbe"))

        elif (lexeme is None): # add info so the lists are same size as batch size
            l_prs_1sg.append("unk")
            l_prs_2sg.append("unk")
            l_prs_3sg.append("unk")
            l_prs_1pl.append("unk")
            l_prs_2pl.append("unk")
            l_prs_3pl.append("unk")
            
            l_ipfv_1sg.append("unk")
            l_ipfv_2sg.append("unk")
            l_ipfv_3sg.append("unk")
            l_ipfv_1pl.append("unk")
            l_ipfv_2pl.append("unk")
            l_ipfv_3pl.append("unk")
        
            l_fut_1sg.append("unk")
            l_fut_2sg.append("unk")
            l_fut_3sg.append("unk")
            l_fut_1pl.append("unk")
            l_fut_2pl.append("unk")
            l_fut_3pl.append("unk")
        
            l_cond_1sg.append("unk")
            l_cond_2sg.append("unk")
            l_cond_3sg.append("unk")
            l_cond_1pl.append("unk")
            l_cond_2pl.append("unk")
            l_cond_3pl.append("unk")
        
            l_sbjv_1sg.append("unk")
            l_sbjv_2sg.append("unk")
            l_sbjv_3sg.append("unk")
            l_sbjv_1pl.append("unk")
            l_sbjv_2pl.append("unk")
            l_sbjv_3pl.append("unk")
        
            l_pst_1sg.append("unk")
            l_pst_2sg.append("unk")
            l_pst_3sg.append("unk")
            l_pst_1pl.append("unk")
            l_pst_2pl.append("unk")
            l_pst_3pl.append("unk")
        
            l_pst_sbjv_1sg.append("unk")
            l_pst_sbjv_2sg.append("unk")
            l_pst_sbjv_3sg.append("unk")
            l_pst_sbjv_1pl.append("unk")
            l_pst_sbjv_2pl.append("unk")
            l_pst_sbjv_3pl.append("unk")
        
            l_imp_2sg.append("unk")
            l_imp_1pl.append("unk")
            l_imp_2pl.append("unk")
            
    return {
    "prs.1sg":l_prs_1sg,
    "prs.2sg":l_prs_2sg,
    "prs.3sg":l_prs_3sg,
    "prs.1pl":l_prs_1pl,
    "prs.2pl":l_prs_2pl,
    "prs.3pl":l_prs_3pl,

    "ipfv.1sg":l_ipfv_1sg,
    "ipfv.2sg":l_ipfv_2sg,
    "ipfv.3sg":l_ipfv_3sg,
    "ipfv.1pl":l_ipfv_1pl,
    "ipfv.2pl":l_ipfv_2pl,
    "ipfv.3pl":l_ipfv_3pl,

    "fut.1sg":l_fut_1sg,
    "fut.2sg":l_fut_2sg,
    "fut.3sg":l_fut_3sg,
    "fut.1pl":l_fut_1pl,
    "fut.2pl":l_fut_2pl,
    "fut.3pl":l_fut_3pl,

    "cond.1sg":l_cond_1sg, 
    "cond.2sg":l_cond_2sg,
    "cond.3sg":l_cond_3sg,
    "cond.1pl":l_cond_1pl,
    "cond.2pl":l_cond_2pl,
    "cond.3pl":l_cond_3pl,

    "sbjv.1sg":l_sbjv_1sg,
    "sbjv.2sg":l_sbjv_2sg,
    "sbjv.3sg":l_sbjv_3sg,
    "sbjv.1pl":l_sbjv_1pl,
    "sbjv.2pl":l_sbjv_2pl,
    "sbjv.3pl":l_sbjv_3pl,

    "pst.1sg":l_pst_1sg,
    "pst.2sg":l_pst_2sg,
    "pst.3sg":l_pst_3sg,
    "pst.1pl":l_pst_1pl,
    "pst.2pl":l_pst_2pl,
    "pst.3pl":l_pst_3pl,

    "pst.sbjv.1sg":l_pst_sbjv_1sg,
    "pst.sbjv.2sg":l_pst_sbjv_2sg,
    "pst.sbjv.3sg":l_pst_sbjv_3sg,
    "pst.sbjv.1pl":l_pst_sbjv_1pl,
    "pst.sbjv.2pl":l_pst_sbjv_2pl,
    "pst.sbjv.3pl":l_pst_sbjv_3pl,

    "imp.2sg":l_imp_2sg,
    "imp.1pl":l_imp_1pl,
    "imp.2pl":l_imp_2pl,
    }

df = dataset.map(get_ortho, batched=True, batch_size=2, num_proc=8, remove_columns=["inf", "prs.ptcp", "pst.ptcp.m.sg", "pst.ptcp.m.pl", "pst.ptcp.f.sg", "pst.ptcp.f.pl"]) # participe passé pas dans classe
```
